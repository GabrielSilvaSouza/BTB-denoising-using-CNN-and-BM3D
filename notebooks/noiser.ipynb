```python
import torch
import matplotlib.pyplot as plt
from PIL import Image
import torchvision.transforms.functional as TF
from torchvision import transforms
```

```python
sigma = 0.07

path_file = '/home/mint/Downloads/img/color/30.jpg'
clean = Image.open(path_file).convert("RGB")

transform = transforms.ToTensor()
clean = transform(clean)

noise = torch.randn_like(clean) * sigma
noisy_tensor = clean + noise

noisy_tensor = torch.clamp(noisy_tensor, 0, 1)

noisy = transforms.ToPILImage()(noisy_tensor)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(clean.permute(1, 2, 0))  # clean image
axes[0].set_title('Clean Image')
axes[0].axis('off')

axes[1].imshow(noisy)  # noisy image
axes[1].set_title('Noisy Image (Sigma = 0.1)')
axes[1].axis('off')

plt.show()
```

```python
noisy.save('noisy_image.png')
```
